```python
# Add dependencies
import glob
import pandas as pd
```

Why use glob?
1. Allows to read a list of .csv files and returns a list(str)
2. Allows pd.read_csv(f) returns Pandas DataFrame
3. pd.concat() returns one Pandas DataFrame

```python
# Read in files for 2019
data19 = pd.concat([pd.read_csv(f) for f in glob.glob('Data/2019/*.csv')], ignore_index = True)
data19.head()
```

```python
data19.count()
```

```python
# Read in files for 2020
data20 = pd.concat([pd.read_csv(f) for f in glob.glob('Data/2020/*.csv')], ignore_index = True)
data20.head()
```

```python
data20.count()
```

```python
# Read in files for 2021
data21 = pd.concat([pd.read_csv(f) for f in glob.glob('Data/2021/*.csv')], ignore_index = True)
data21.head()
```

```python
data21.count()
```

```python
# Export each back to Data File as a complete csv
data19.to_csv("TableauData/data19.csv")
data20.to_csv("TableauData/data20.csv")
data21.to_csv("TableauData/data21.csv")
```
